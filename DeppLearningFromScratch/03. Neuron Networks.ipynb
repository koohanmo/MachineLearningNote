- step_function : 입력이 0을 넘으면 1 아니면 0
- sigmoid function : 1/1+exp(-x)
- ReLU function : 입력이 0을 넘으면 x, 아니면 0 <-- 최근에 많이사용

```python
def step_function(x):
    if x>0: return 1
    else :return 0
```

- 위의 step_function은 실수 1개만 가능하고, numpy의 Array를 지원하지않는다

```python
import numpy as np
def step_function(x):
    y= x>0   #모든 원소가 0보다 큰지 보고 True/False로 변환
    return y.astype(np.int) #원소를 True==1, False=0으로 변환

#한줄로도 표현 할 수 있다
def step_function(x):return np.array(x>0,dtype=np.int)
```

##  구현 정리
- 신경망 구현 관례에 따라 가중치만 대문자(W1,W2..) 나머지 Bias와 중간결과(A,Z)는 모두 소문자

```python
def init_network():
    """
    가중치와 편향을 초기화 해서 딕셔너리에 저장
    """
    network={}
    network['W1'] = np.array([[.1,.3,.5],[.2,.4,.6]])
    network['b1'] = np.array([.1,.2,.3])
    network['W2'] = np.array([[.1,.4],[.2,.4],[.3,.6]])
    network['b2'] = np.array([.1,.2])
    network['W3'] = np.array([[.1,.3],[.2,.4]])
    network['b3'] = np.array([.1,.2])
    return network

def forward(network, x):
    """
    신호가 순방향으로 전달(순전파)
    """
    W1,W2,W3 = network['W1'],network['W2'],network['W3']
    b1,b2,b3 = network['b1'],network['b2'],network['b3']
    
    a1= np.dot(x,W1)+b1
    z1=sigmoid(a1)
    a2 = np.dot(z1,W2)+b2
    z2=sigmoid(a2)
    a3= np.dot(z2,W3)+b3
    y=identify_function(a3)
    
    return y
```

```python
network = init_network()
x= np.array([1.0,0.5])
y= forward(network,x)
print (y)
```

####  소프트 맥스 함수(softmaax function)
- 분류에서 사용
- Yk = exp(ak)/Sigma(exp(ai))

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y= exp_a/sum_exp_a
    return y
```

###  소프트맥스는 지수함수라 오버플로우에 주의해야한다.
- 분모/분자에 C를 곱해도 무관
- 이것은 exp()안에 각각 들어가면 logC를 더한 것과 같다.
- 결과적으로 분자/분모의 exp()안에 같은수 C를 더하거나 빼도 결과는 같다는 소리다
- 주로 C'로 입력신호중 최대값을 이용한다.

```python
a = np.array([1010,1000,990])
print (np.exp(a)/np.sum(np.exp(a)))
```

```python
c=np.max(a)
print (np.exp(a-c)/np.sum(np.exp(a-c)))
```

```python
def softmax(a):
    c= np.max(a)
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y=exp_a/sum_exp_a
    return y
```

###  소프트맥스의 출력은 [0,1.0]의 실수이고 출력의 총합은 1이다 => 확률로 해석 할 수 있다.
### 또한 exp(x)가 단조증가 함수이기 떄문에 소프트맥스를 적용해도 대소는 변하지 않는다 -> 마지막층에서 생략 가능
### 실제 학습시킬 때에는 출력층에 소프트맥스를 사용하고, 추론할 때(현장)에서는 사용하지 않는다

```python
import sys, os
import numpy as np
from dataset.mnist import load_mnist

#훈련이미지, 훈련레이블, 시험이미지, 시험 레이블
(x_train, t_train), (x_test, t_test) = load_mnist(flatten=True, normalize=False)
"""
normalize : 이미지 픽셀의 값을 [0.0,1.1]으로 정규화
flatten : 이미지를 1차원 배열로
one_hot_label : 원-핫 인코딩
"""

print(x_train.shape)
print(t_train.shape)
print(x_test.shape)
print(t_test.shape)
```

```python
from PIL import Image


def img_show(img):
    pil_img = Image.fromarray(np.uint8(img))
    pil_img.show()

(x_train, t_train), (x_test, t_test) = load_mnist(flatten=True, normalize=False)

img = x_train[0]
label = t_train[0]
print(label)  # 5

print(img.shape)  # (784,)
img = img.reshape(28, 28)  # 形状を元の画像サイズに変形
print(img.shape)  # (28, 28)

img_show(img)
```

```python
def get_data():
    (x_train,t_train),(x_test,t_test) =load_mnist(normalize=True,flatten=True,one_hot_label=False)
    return x_test, t_test

import pickle
def init_network():
    with open("sample_weight.pkl",'rb') as f:
        network= pickle.load(f)
    return network
def predict(network, x):
    W1,W2,W3 = network['W1'], network['W2'], network['W3']
    b1,b2,b3 = network['b1'], network['b2'], network['b3']
    
    a1 = np.dot(x,W1)+b1
    z1=sigmoid(a1)
    a2 = np.dot(z1,W2)+b2
    Z2= sigmoid(a2)
    a3 = np.dot(Z2,W3)+b3
    y=softmax(a3)
    
    return y
```

```python
x,t = get_data()
network = init_network()

accuracy_cnt =0
for i in range(len(x)):
    y = predict(network,x[i])
    p = np.argmax(y)
    if p == t[i] : accuracy_cnt +=1

print ("정확도 : {}".format(float(accuracy_cnt)/len(x)))
```

### 이미지를 1장씩 처리를 하는 것보다 Batch로 처리하는 것이더 빠르다.

```python
x,t = get_data()
network = init_network()

batch_size=100
accuracy_cnt =0
for i in range(0,len(x),batch_size):
    x_batch = x[i:i+batch_size]
    y_batch = predict(network,x_batch)
    p = np.argmax(y_batch,axis=1) #1차원 측으로 최대값
    
    accuracy_cnt += np.sum(p==t[i:i+batch_size])

print ("정확도 : {}".format(float(accuracy_cnt)/len(x)))
```
