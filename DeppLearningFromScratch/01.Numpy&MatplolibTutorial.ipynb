#####  산술 연산
##### Element 끼리 + - / *, 원소수가 다르면 에러

```python
x= np.array([1,2,3])
y= np.array([2,4,6])
print(x+y)
print(x-y)
print(x/y)
print(x*y)
```

```python
x=np.array([1,2])
print (x+y)
```

scalar value를 브로드캐스트

```python
x= np.array([1.0,2.0,3.0])
print (x/2.0)
```

#####  Numpy의 N차원 배열

```python
A=np.array([[1,2],[3,4]])
print(A,'\n',A.shape,A.dtype)
```

```python
B=np.array([[3,0],[0,6]])
print (A+B)
print (A*B)
# 원소끼리의 곱  
```

###  참고 : 1차원=Vector, 2차원=Matrix, 3차원=Tensor

#####  브로드 캐스트(Broadcast)
#####  Numpy에서 다른 형상의 배열끼리 계산하기 위해서 자동으로 행렬을 확대시켜줌

```python
A= np.array([[1,2],[3,4]])
B=np.array([10,20])
print (A*B)
```

A*B를 할 때에 B가 2차원 배열로 확대됨 [10,20][10,20]

#####  원소 접근

```python
X=np.array([[51,55],[14,19],[0,4]])
print (X)
print (X[0])     # 행
print (X[0][1])   # i,j 원소
```

```python
for row in X:
    for col in row:
        print(col,end=" ")
    print()
```

```python
X= X.flatten() #X를 1차원으로 평탄화
X
```

##  matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x= np.arange(0,6,0.1)
y =np.sin(x)
print (y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
#데이터 준비
x= np.arange(0,6,0.1)
y1=np.sin(x)
y2=np.cos(x)
```

```python
#그래프
plt.plot(x,y1,label="sin")
plt.plot(x,y2,linestyle="--",label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin&cos')
plt.legend()
plt.show()
```
